```python
import os
import json
import sys
import random
import math

from gensim.models import Word2Vec
from nltk.cluster import KMeansClusterer
import nltk
from sklearn import cluster
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
from matplotlib import pyplot as plt
from yellowbrick.cluster import KElbowVisualizer

%matplotlib inline
```

```python
_dir_path = os.path.dirname(os.path.abspath(''))
modelVocab = []
modelVectors = []
```

```python
def read_vectors():
    d2v_vectors = json.load(open(os.path.join(_dir_path , 'data', 'sequence_vectors_doc2vec.json')))
    d2v_vectors = d2v_vectors['vectors']
    
    for i, vector in enumerate(d2v_vectors):
        modelVocab.append(vector)
        modelVectors.append(d2v_vectors[vector])
```

```python
read_vectors()
```

```python
print(modelVocab[0])
```

```python
k_rng = range(1,100)
sse = []

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(modelVectors)
    sse.append(km.inertia_)
```

```python
sse[0]
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse)
```

```python
k_rng1 = range(1,10)
sse1 = []

for k in k_rng1:
    km = KMeans(n_clusters=k)
    km.fit(modelVectors)
    sse1.append(km.inertia_)
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng1, sse1)
```

```python
k_rng2 = range(1,15)
sse2 = []

for k in k_rng2:
    km = KMeans(n_clusters=k)
    km.fit(modelVectors)
    sse2.append(km.inertia_)
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng2, sse2)
```

```python
k_rng3 = range(1,30)
sse3 = []

for k in k_rng3:
    km = KMeans(n_clusters=k)
    km.fit(modelVectors)
    sse3.append(km.inertia_)
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng3, sse3)
```

```python
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,500))

visualizer.fit(np.array(modelVectors))       # Fit the data to the visualizer
visualizer.show()                            # Finalize and render the figure
```
